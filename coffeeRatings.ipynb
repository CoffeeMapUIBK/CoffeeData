```python
import pandas as pd
```

```python
arabica = pd.read_csv("./arabica_data_cleaned.csv")
```

```python
arabica
```

```python
arabica['Country.of.Origin'] = arabica['Country.of.Origin'].str.strip().str.title()
arabica['Harvest.Year'] = arabica['Harvest.Year'].str.strip()
```

```python
# Replace any known anomalies in country names, e.g., 'Usa' to 'USA'
arabica['Country.of.Origin'] = arabica['Country.of.Origin'].replace({'Usa': 'USA', 'Uk': 'United Kingdom'})  # Add more replacements as needed
```

```python
arabica.dropna(subset=['Country.of.Origin', 'Harvest.Year'], inplace=True)
```

```python
# Convert Harvest Year to int if necessary, checking if there are non-numeric values first
arabica = arabica[pd.to_numeric(arabica['Harvest.Year'], errors='coerce').notna()]  # Removes non-numeric years
arabica['Harvest.Year'] = arabica['Harvest.Year'].astype(int)
```

```python
ratings_columns = ['Aroma', 'Flavor', 'Aftertaste', 'Acidity', 'Body', 'Balance', 'Uniformity', 'Clean.Cup', 'Sweetness', 'Cupper.Points', 'Total.Cup.Points']
```

```python
mean_ratings = arabica.groupby(['Country.of.Origin', 'Harvest.Year'])[ratings_columns].mean().reset_index()
```

```python
mean_ratings
```

```python
mean_ratings.to_csv('processed_coffee_ratings_arabica.csv', index=False)
```
